# Menghitung Persamaan regresi

```python
from scipy.special import stdtrit
from tabulate import tabulate
import numpy as np
import math

pendapatan = [8,12,16,20,24,26]
konsumsi = [7,9,12,14,13,15]

#banyaknya data
n = len(pendapatan)

#Xi = Pendapatan^2
Xi = pow(np.array(pendapatan),2)

#Yi = Konsumsi^2
Yi = pow(np.array(konsumsi),2)

#XiYi = pendapatan * konsumsi
XiYi = np.array(pendapatan) * konsumsi

x = np.sum(pendapatan) / n
y = np.sum(konsumsi) / n

b = ((n * np.sum(XiYi)) - (np.sum(pendapatan) * np.sum(konsumsi))) / ((n * np.sum(Xi)) - pow(np.sum(pendapatan),2))
a = y - (b * x)

# contoh kasus pendapatan karyawan sebesar 23 juta
pendapatanKaryawan = 23
besaranKonsumsi = a + b * pendapatanKaryawan 

#statistik uji pengaruh pendapatan terhadap konsumsi
alpha = 5/100
titik_kritis = stdtrit((n - 2), 1 - alpha/2)
t_table = float("%.3f" % titik_kritis)
daerah_kritis = np.array([np.negative(t_table), t_table])

# uji pendapatan berpengaruh terhadap konsumsi
penduga_standar_deviasi = math.sqrt(((np.sum(Yi) - (a * np.sum(konsumsi)) ) - (b * np.sum(XiYi))) / (n - 2))
penduga_kesalahan_baku = pow(penduga_standar_deviasi, 2) / (np.sum(Xi) - pow(np.sum(pendapatan),2) / n)
t_hit = b - 0 / penduga_kesalahan_baku

# uji korelasi pendapatan terhadap konsumsi
#hitung nilai determinasi terlebih dahulu
R = (pow((n * np.sum(XiYi)) - (np.sum(pendapatan)*np.sum(konsumsi)),2)) / ((n * np.sum(Xi) - pow(np.sum(pendapatan),2)) * (n * np.sum(Yi) - pow(np.sum(konsumsi), 2)))

#statistik uji korelasi
taraf_nyata = 5/100
titik_kritis_korelasi = stdtrit((n - 2), 1 - taraf_nyata/2)
t_table_korelasi = float("%.3f" % titik_kritis_korelasi)
daerah_kritis_korelasi = np.array([np.negative(t_table_korelasi), t_table_korelasi])

#menetukan koefisien korelasi
r = math.sqrt(R)
t_hit_korelasi = (r * math.sqrt(n - 2)) / (math.sqrt(1 - R))
```

## Hasil

```python
table = [["Pendapatan",str(np.sum(pendapatan))],["Konsumsi",str(np.sum(konsumsi))],["Xi^2",str(np.sum(Xi))],["Yi^2",str(np.sum(Yi))],["XiYi",str(np.sum(XiYi))]]

print(tabulate([["x",str(x)],["y", str(y)],["a",str(a)],["b",str(b)],["R^2",str(R)],["r",str(r)]],["Variabel","Nilai"],tablefmt="fancy_grid"))

print(tabulate({"Pendapatan": pendapatan,"Konsumsi": konsumsi, "Xi^2": Xi, "Yi^2": Yi, "XiYi": XiYi}, headers="keys", showindex="always",tablefmt="fancy_grid"))
print(tabulate(table, ["Nama tabel","Jumlah"],tablefmt="fancy_grid"))

print("Hasil hipotesis pengaruh pendapatan")
if(daerah_kritis[0] > t_hit > daerah_kritis[1] and t_hit != 0):
    print("t = %.2f" % t_hit,':',"Ada pengaruh antara pendapatan dengan konsumsi rumah tangga")
else:
    print("t = %.2f" % t_hit,':',"Tidak ada pengaruh antara pendapatan dengan konsumsi rumah tangga")

print("\nHasil hipotesis korelasi")
if(daerah_kritis_korelasi[0] > t_hit > daerah_kritis_korelasi[1] and t_hit != 0):
    print("t = %.2f" % t_hit_korelasi,':',"Ada korelasi antara pendapatan dengan konsumsi rumah tangga")
else:
    print("t = %.2f" % t_hit_korelasi,':',"Tidak ada korelasi antara pendapatan dengan konsumsi rumah tangga")

print(titik_kritis)
```
